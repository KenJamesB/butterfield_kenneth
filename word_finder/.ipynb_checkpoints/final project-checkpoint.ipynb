```python
# Start your code here. You may make as many code cells as you need.

import re
import pprint
pp = pprint.PrettyPrinter()
import pathlib
import nltk
from nltk.tokenize import sent_tokenize
from nltk.tokenize import RegexpTokenizer      # Import RegexpTokenizer class (type)
tokenizer = RegexpTokenizer("['\w']+")


def word_list(file):                           # Creates a list of words I want to find based on a .txt file.
    
    fhandle = open(file, errors = 'ignore')
    wordlist = fhandle.read()
    wordlist = wordlist.split()
    return wordlist


def number_sentences(file):                    # This program numbers all sentences in a file. I don't need to use
                                               # it again, since I've used it on the texts already.
    fhandle = open(file, errors = 'ignore')
    text_as_string = fhandle.read()
    sentences = sent_tokenize(text_as_string)
    fhandle.close()
    
    n = 1
    fhand = open(file, 'w')
    for sent in sentences:
        fhand.write('({}) '.format(n) + sent + ' ')
        n += 1
        
    fhand.close()
    
    
def corpus_number_sentences(folder):           
   
    folder_path = pathlib.Path(folder)         # Make a path object for the folder 'folder'
    files = folder_path.glob('*.txt')          # Find all the files in the folder
    for file_path in files:                    # For each file,
        number_sentences(file_path)            # Runs the file through number_sentences().
                                               # I have already run this function, so I shouldn't run it again, or
                                               # the sentences will be numbered twice.

def word_finder(words, text):                  # Extracts sentences from a string that contain words from the word-
                                               # list created by word_list.
    wordlist = word_list(words)
    matches = list()
    
    fhandle = open(text, errors = 'ignore')
    text_as_string = fhandle.read()
    sentences = sent_tokenize(text_as_string)
    
    for sent in sentences:                     # Populates a list of matched sentences.
        sent = sent.strip('.?!')
        sent = sent.lower()
        words = sent.split()
        for word in wordlist:
            for item in words:
                if word == item:
                    if sent not in matches:
                        matches.append(sent)
                        
    results = len(matches)
    matches.append(results)
                        
    return matches


def corpus_word_finder(wordlist, folder):
   
    titles = dict()
    
    folder_path = pathlib.Path(folder)         # Make a path object for the folder 'folder'
    files = folder_path.glob('*.txt')          # Find all the files in the folder
    
    for file_path in files:                    # For each file,
        titles[str(file_path)] = word_finder(wordlist, file_path)       # Runs the file through word_finder().
        
    return titles
    

def find_before_patterns(words, fhand, n):     # Finds the 'n' words that occur before the words in the list in
                                               # a file.
    """ 
    text: paragraphs/sentences etc.
    n:    now many words before the keywords that should be included in the results
    """
    text = (open(fhand)).read()
    lst = word_list(words)
    result = list()
    sentences = sent_tokenize(text)
    
    for word in lst:
        pattern = "\w+\s"*n + word 
        for sent in sentences:
            matches = re.findall(pattern, sent) 
            if matches:
                result.extend(matches)
                
    return result


def find_adj_patterns(lst):                    # Finds adjectives/modifiers prior to nature words using
                                               # find_before_patterns().
    
    parts = ['JJ', 'JJR', 'JJS', 'RB', 'RBR', 'RBS']
    adj = list()
    
    for item in lst:
        text = tokenizer.tokenize(item)
        pos = nltk.pos_tag(text)
        for part in parts:
            pair = pos[0]
            if pair[1] == part:
                adj.append(item)
            
    return adj


def corpus_adj_patterns(folder):
    
    adjectives = dict()
    
    folder_path = pathlib.Path(folder)         # Make a path object for the folder 'folder'
    files = folder_path.glob('*.txt')          # Find all the files in the folder
    
    for file_path in files:                    # For each file,
        lst = find_before_patterns('word_list.txt', file_path, 1)
        adjectives[str(file_path)] = find_adj_patterns(lst)       # Runs the file through find_adj_patterns().
        
    return adjectives


pp.pprint(corpus_adj_patterns('Emerson'))
pp.pprint(corpus_adj_patterns('Hawthorne'))


# The commands below will extract all sentences that contain a nature word, in case you are curious to see them.
# The results produced by the sentences informed my report, but I commented them out here since there are many.

# pp.pprint(corpus_word_finder('word_list.txt', 'Emerson'))
# pp.pprint(corpus_word_finder('word_list.txt', 'Hawthorne'))
```
